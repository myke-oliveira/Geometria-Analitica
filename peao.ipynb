```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

u = np.linspace(0, 2*np.pi, 30)
v = np.linspace(0, 2, 30)
u, v = np.meshgrid(u, v)
x = 10*np.cos(u); y = 10*np.sin(u); z = v
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, alpha=.6, cmap='bone')

x = 10*np.cos(u)*np.cos(v); y = 10*np.sin(u)*np.cos(v); z = 2 + 0*u;
ax.plot_surface(x, y, z, alpha=.6, cmap='bone')

x = 10*np.cos(u)*np.cos(u); y = 10*np.sin(u)*np.cos(v); z=0*u;
ax.plot_surface(x, y, z, alpha=.6, cmap='bone')

x = 8*np.cos(u)*np.cos(v); y=8*np.sin(u)*np.cos(v); z = 4*np.sin(v);
ax.plot_surface(x, y, z+2, alpha=.6, cmap='bone')

u = np.linspace(0, 2*np.pi, 30)
v = np.linspace(0, 1.5, 30)
u, v = np.meshgrid(u, v)
x = 2*np.cos(u)*np.cosh(v); y = 2*np.sin(u)*np.cosh(v); z = -4*np.sinh(v);
ax.plot_surface(x, y, z+14, alpha=.6, cmap='bone')

u = np.linspace(0, 2*np.pi, 30)
v = np.linspace(0, 2*np.pi, 30)
u, v = np.meshgrid(u, v)
x = 4*np.cos(u)*np.cos(v); y = 4*np.sin(u)*np.cos(v); z = 4*np.sin(v);
ax.plot_surface(x, y, z+16, alpha=.6, cmap='bone')

x = 6*np.cos(u)*np.cos(v); y = 6*np.sin(u)*np.cos(v); z = 1*np.sin(v);
ax.plot_surface(x, y, z+6, alpha=.6, cmap='bone')

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
plt.show()
```
