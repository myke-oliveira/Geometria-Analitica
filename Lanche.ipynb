# Fazendo um lanche com Python e Geometria Anal√≠tica

```python
# bibliotecas
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,12))

# Copo
# Base do Copo
# Raio da base 3cm
u = np.linspace(0, 2*np.pi, 30)
v = np.linspace(0, 3, 30)
u, v = np.meshgrid(u, v)
x = v*np.cos(u); y = v*np.sin(u); z = 0*u;
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, alpha=0.6, cmap='bone')

# Lateral do Copo
# Base inferior 3cm
# Base superior 3.5cm
# Altura 10cm
u = np.linspace(0, 2*np.pi, 30)
v = np.linspace(3, 3.5, 30)
u, v = np.meshgrid(u, v)
x = v*np.cos(u); y = v*np.sin(u); z = (v - 3) * 20;
ax.plot_surface(x, y, z, alpha=0.6, cmap='bone')

# Tampa do Copo
# Raio da Tampa 3.5cm
# Raio do Furo 0.5cm
u = np.linspace(0, 2*np.pi, 30)
v = np.linspace(0.5, 3.5, 30)
u, v = np.meshgrid(u, v)
x = v*np.cos(u); y = v*np.sin(u); z = 0*u + 10;
ax.plot_surface(x, y, z, alpha=0.6, cmap='bone')

# Canudo
# Raio do Canudo 0.4cm
# Altura do Canudo 13cm
u = np.linspace(0, 2*np.pi, 30)
v = np.linspace(0, 13)
u, v = np.meshgrid(u, v)
x = 0.4 * np.cos(u); y = 0.4 * np.sin(u); z = v + 2;
ax.plot_surface(x, y, z, alpha=0.6, cmap='bone')

plt.show()
```
