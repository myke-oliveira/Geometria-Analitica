```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint

a = randint(1, 11)
b = randint(1, 11)
m_1 = randint(1, 11)
m_2 = randint(1, 11)
n_1 = randint(1, 11)
n_2 = randint(1, 11)
n_3 = randint(1, 11)

theta_max = 2*np.pi

theta = np.linspace(0, theta_max, 100)

r = (np.abs(np.cos(m_1*theta/4)/a)**n_2 + np.abs(np.sin(m_2*theta/4)/b)**n_3)**(-1/n_1)

print(a, b, m_1, m_2, n_1, n_2, n_3)

plt.polar(theta, r)
plt.show()
```
